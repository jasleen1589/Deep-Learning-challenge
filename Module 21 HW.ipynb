```python
#  Import and read the charity_data.csv.
import pandas as pd
import tensorflow as tf
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=["EIN", "NAME"])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df.APPLICATION_TYPE.value_counts()
application_type_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(application_type_counts[application_type_counts < 500].index)


# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df.CLASSIFICATION.value_counts()
classification_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts[classification_counts > 1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(classification_counts[classification_counts < 1000].index)


# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"].values
X = application_df.drop(columns=["IS_SUCCESSFUL"]).values


# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Assuming initial setup and preprocessing steps are done

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Adjust model structure as per optimization strategy
model = Sequential()
model.add(Dense(units=X_train_scaled.shape[1], activation='relu'))

# Adjust the number of neurons
model.add(Dense(units=32, activation='relu'))

# Add more layers or adjust neurons as needed
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Callback to save model's weights every 5 epochs
checkpoint = tf.keras.callbacks.ModelCheckpoint(
    'weights.{epoch:02d}.hdf5',
    save_weights_only=True,
    period=5
)

# Train the model with adjusted epochs and batch size
adjusted_epochs = 100
adjusted_batch_size = 32
history = model.fit(
    X_train_scaled, y_train,
    epochs=adjusted_epochs,
    batch_size=adjusted_batch_size,
    callbacks=[checkpoint],
    validation_data=(X_test_scaled, y_test)
)
```

```python
# Add a checkpoint to save weights every 5 epochs
checkpoint = ModelCheckpoint("weights.best.hdf5", verbose=1, save_best_only=True, save_weights_only=True, period=5)

# Train the model, adjust epochs or batch size as part of optimization
model.fit(X_train_scaled, y_train, epochs= int(adjusted_epochs), batch_size=int(adjusted_batch_size), callbacks=[checkpoint], validation_data=(X_test_scaled, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test_scaled, y_test)
print(f"Loss: {loss}, Accuracy: {accuracy}")

# Save the optimized model
model.save('AlphabetSoupCharity_Optimization.h5')

print(type(adjusted_epochs))
print(type(adjusted_batch_size))
```

```python
import tensorflow as tf

# Define the model - deep neural net, the number of input features and hidden nodes for each layer.
number_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 30

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))


# Assuming a binary classification problem, using 1 unit with 'sigmoid' activation function
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Define the number of epochs
epochs = 100

# Train the model
history = nn.fit(
    X_train_scaled,
    y_train,
    epochs=epochs,
    batch_size=32,
    validation_data=(X_test_scaled, y_test),
    verbose=1
)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
